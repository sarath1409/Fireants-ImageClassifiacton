```python
import os, shutil
import numpy as np
from skimage import data, img_as_float, exposure
import random
from keras.preprocessing.image import ImageDataGenerator,load_img, img_to_array, array_to_img

def change_gamma(image):
    if np.random.random_sample() > 0.5:
        return exposure.equalize_hist(image)
    else:
        return exposure.adjust_gamma(image, 1+np.random.random_sample())

datagen = ImageDataGenerator(
    featurewise_center=False,
     samplewise_center=True,
     featurewise_std_normalization=False,
     samplewise_std_normalization=False,
     zca_whitening=False,
     zca_epsilon=True,
     rotation_range=5,
     width_shift_range=0,
     height_shift_range=0,
     shear_range=0,
     zoom_range=0,
     channel_shift_range=1,
     fill_mode='nearest',
     cval=0.,
     horizontal_flip=True,
     vertical_flip=False,
     rescale=1/255.,
     preprocessing_function=change_gamma
)


base_dir = 'FireAnt/'
dir_names = os.listdir(base_dir)
for i in os.listdir(base_dir):
    print(i)
    count=0
    for j in os.listdir(os.path.join(base_dir,i)):
        num=0
        img = load_img(os.path.join(base_dir,i,j))
	inputFilepath =os.path.join(base_dir,i,j)
	filename_w_ext = os.path.basename(inputFilepath)
	filename, file_extension = os.path.splitext(filename_w_ext)
        x = img_to_array(img)
        x = x.reshape((1,)+x.shape)
        for batch in datagen.flow(x, batch_size=1, save_to_dir=base_dir+i, save_prefix=filename, save_format='jpg'):
            num += 1
            if num ==1:
                break
            print('Augmenting ' + j + ' : iter-',(num))
        count += 1
        print(count)
```

```python
# Import
import os
import shutil

# Set the directory of the folder containing the images
# This root directory MUST contain severla sub-folders 
# This sub-folder MUST contain several images of one class
# Eg: This is how the folder structure should look like
# root-folder IS THE image_dir
#     root-folder > class-1 > *.jpg 
#				  > class-2 > *.jpg
#				  > ... 
#				  > class-N > *.jpg
# Please make sure that ther are no additional folders in the sub folders
# or in the images folders
image_dir = "/media/sarath/Stuff/Coding/Siya/FireAnt"

# Change the percentage of images in the sub-folder you want
# to be assigned to the training and validation set.
# The percentage of images in the testing set will be calculated
# based on the supplied values
training_percentage = 0.6
validation_percentage = 0.2

# The OUTPUT-FOLDER will be created one level above the IMAGE_DIR(in its parent directory)
# From terminal execute a 'cd ..' to see the output folder named: 'output- folder' 

# PLEASE DONT EDIT ANYTHING BELOW THIS LINE
print('Image directory provided:' + image_dir)
print()

parent_dir = os.path.dirname(os.path.dirname(image_dir))
output_dir = parent_dir + "/output-folder/"
training_dir = output_dir + "training/"
test_dir = output_dir + "test/"
validation_dir = output_dir + "validation/"

# check if the output folder exists in the parent directory
# If not make it. 
if os.path.exists(output_dir):
	print("A directory with the name 'output-folder' exists.")
	print("Deleting the folder...")
	shutil.rmtree(output_dir)
	print()

print("Creating a directory named 'output-folder' to store output...")
os.mkdir(output_dir)
# Create sub-folders for the training, test and validation images
print('Creating sub-folders to store training, validation and test set results..')
print()
os.mkdir(training_dir)
os.mkdir(validation_dir)
os.mkdir(test_dir)

list_dirs = [x for x in os.listdir(image_dir) if os.path.isdir(os.path.join(image_dir, x))]

for dirs in list_dirs:
	print("Working on class:" + dirs)
	list_files = [x for x in os.listdir(os.path.join(image_dir, dirs))]
	no_files = len(list_files)

	no_training_set_images = int(no_files * training_percentage)
	no_validation_set_images = int(no_files * validation_percentage)
	no_testing_set_images = no_files - no_training_set_images - no_validation_set_images

	training_files = list_files[:no_training_set_images]
	validation_files = list_files[no_training_set_images:no_training_set_images + no_validation_set_images]
	testing_files = list_files[-no_testing_set_images:]

	class_training_dir = training_dir + dirs
	os.mkdir(training_dir + dirs)
	for file in training_files:
		shutil.copy(os.path.join(image_dir, dirs, file), class_training_dir)

	class_validation_dir = validation_dir + dirs
	os.mkdir(class_validation_dir)
	for file in validation_files:
		shutil.copy(os.path.join(image_dir, dirs, file), class_validation_dir)

	class_test_dir = test_dir + dirs
	os.mkdir(class_test_dir)
	for file in testing_files:
		shutil.copy(os.path.join(image_dir, dirs, file), class_test_dir)

	print("Created training, test, and validation sub-folder for the class:'" + dirs + "'")
	print()
```
